```python
# Import necessary libraries
import zipfile  # For working with ZIP files
import os  # For file and directory operations
import pandas as pd  # For data manipulation and analysis

# Define the path to the ZIP file and the directory to extract its contents
zip_file_path = "Ecommerce Purchases.zip"  # Path to the ZIP file
extract_to = "./extracted_data"  # Directory where files will be extracted

# Open the ZIP file in read mode and extract its contents
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_to)  # Extract all files to the specified directory

print("File extracted successfully!")  # Notify the user that the extraction is complete

# List all files in the extraction directory
extracted_files = os.listdir(extract_to)  # Get a list of all files in the directory
print("Extracted files:", extracted_files)  # Print the list of extracted files

# Specify the expected CSV file name and its full path
csv_file_name = 'Ecommerce Purchases.csv'  # The name of the CSV file to look for
csv_file_path = os.path.join(extract_to, csv_file_name)  # Construct the full path to the CSV file

# Check if the CSV file exists in the extracted folder
if os.path.exists(csv_file_path):
    # Read the CSV file into a Pandas DataFrame
    data = pd.read_csv(csv_file_path)  # Load the CSV file into a DataFrame
    print(data.head())  # Print the first 5 rows of the DataFrame for preview
else:
    # Notify the user if the file was not found
    print(f"The file {csv_file_name} was not found in the extracted folder.")  
```
